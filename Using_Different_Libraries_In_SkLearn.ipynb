# one hot encoder

```python
data={'sales': [100000, 222000, 1000000, 522000, 111111, 222222, 1111111, 20000],
      'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonville', 'Miami', 'Jacksonville', 'Miami' , 'Miami'],
      'size': ['Small', 'Medium','Large', 'Large', 'Small', 'Medium', 'Large', 'Small']}
data=pd.DataFrame(data)
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
k=OneHotEncoder(sparse_output=False).set_output(transform="pandas")
```

```python
ki=k.fit_transform(data[["city"]])
```

```python
ki
```

```python
data=pd.concat([data,ki],axis=1).drop(columns=["city"])
```

```python
data
```

# Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data2=data
data2
```

```python
sizes = ['Medium','Large','Small']
ordi = OrdinalEncoder(categories=[sizes])
data2[["size"]] = ordi.fit_transform(data2[["size"]])
data2
```

# simple imputer for numerical data

```python
from sklearn.impute import SimpleImputer
```

```python
ki = pd.DataFrame({'data': [50,62, np.nan,100, 26, 13, 31, 50]})
ki
```

```python
ki.isnull().sum()
```

```python
model=SimpleImputer(strategy="most_frequent",add_indicator=True)
model.fit_transform(ki)
```

```python
modeli=SimpleImputer(strategy="constant",fill_value=123)
modeli.fit_transform(ki)
```

# simple imputer for ordinal data

```python
names = pd.DataFrame({'names': ['ryan', 'nolan', 'honus', 'wagner', np.nan, 'ruth']})
names
```

```python
kii=SimpleImputer(strategy="constant",fill_value="rian",add_indicator=True)
kii.fit_transform(names)
```
